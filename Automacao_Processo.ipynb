### Passo 1 - Importar Arquivos e Bibliotecas

```python
import pandas as pd
import win32com.client as win32
import pathlib
```

### Passo 2 - Leituras dos Arquivos

```python
vendas = pd.read_excel(r'Banco de Dados\Vendas.xlsx')
emails = pd.read_excel(r'Banco de Dados\Emails.xlsx')
lojas = pd.read_csv(r'Banco de Dados\Lojas.csv', encoding='latin1', sep=';')
display(vendas)
display(emails)
display(lojas)
```

```python
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

### Passo 3 - Criar um Dicionário pra cada Loja

```python
dicionario_lojas ={}
for loja in lojas['Loja']:
    dicionario_lojas[loja] = vendas.loc[vendas['Loja'] == loja, :]

#exemplos
display(dicionario_lojas['Iguatemi Campinas'])
display(dicionario_lojas["Center Shopping Uberlândia"])
```

### Passo 4 - Calcular o indicador para 1 loja

```python
dia_indicador = vendas['Data'].max()
print('{}/{}/{}'.format(dia_indicador.day, dia_indicador.month, dia_indicador.year))
```

### Passo 5 - Salvar a planilha na pasta de backup

```python
caminho = pathlib.Path(r'Backup Arquivos Lojas')

arquivos_pasta_backup = caminho.iterdir()
listas_nome_backup = [arquivo.name for arquivo in arquivos_pasta_backup]
for loja in dicionario_lojas:
    if loja not in listas_nome_backup:
        caminho_loja = caminho / loja
        caminho_loja.mkdir()
    nome_arquivo = '{}-{}-{}.xlsx'.format(dia_indicador.day, dia_indicador.month, loja)
    local_arquivo = caminho / loja / nome_arquivo
    dicionario_lojas[loja].to_excel(local_arquivo)
```

### Passo 6 - Enviar E-mail para o gerente de cada Loja

```python
#definição de metas
meta_faturamento_dia = 1000
meta_faturamento_ano = 1650000
meta_qtdeprodutos_dia = 4
meta_qtdeprodutos_ano = 120
meta_ticketmedio_dia = 500
meta_ticketmedio_ano = 500
```

```python
for loja in dicionario_lojas:
    vendas_loja = dicionario_lojas[loja]
    vendas_loja_dia = vendas_loja.loc[vendas_loja['Data']==dia_indicador, :]
    
    faturamento_dia = vendas_loja_dia['Valor Final'].sum()
    faturamento_ano = vendas_loja['Valor Final'].sum()
    
    qtd_produtos_dia = len(vendas_loja_dia['Produto'].unique())
    qtd_produtos_ano = len(vendas_loja['Produto'].unique())
    
    valor_venda = vendas_loja.groupby('Código Venda').sum(numeric_only=True)
    ticket_medio = valor_venda['Valor Final'].mean()
     
    valor_venda_dia = vendas_loja_dia.groupby('Código Venda').sum(numeric_only=True)
    ticket_medio_dia = valor_venda_dia['Valor Final'].mean()
    
    #enviar email
    outlook = win32.Dispatch('outlook.application')
    
    nome = emails.loc[emails['Loja'] == loja, 'Gerente'].values[0]
    mail = outlook.CreateItem(0)
    mail.To = emails.loc[emails['Loja'] == loja, 'E-mail'].values[0]
    mail.Subject = 'Ranking dia {}/{}'.format(dia_indicador.day, dia_indicador.month)
    
    if faturamento_dia >= meta_faturamento_dia:
        cor_fat_dia = 'green'
    else:
        cor_fat_dia = 'red'
    if faturamento_ano >= meta_faturamento_ano:
        cor_fat_ano = 'green'
    else:
        cor_fat_ano = 'red'
    if qtd_produtos_dia >= meta_qtdeprodutos_dia:
        cor_qtd_dia = 'green'
    else:   
        cor_qtd_dia = 'red'
    if qtd_produtos_ano >= meta_qtdeprodutos_ano:
        cor_qtd_ano = 'green'
    else:
        cor_qtd_ano = 'red'
    if ticket_medio_dia >= meta_ticketmedio_dia:
        cor_ticket_dia = 'green'   
    else:   
        cor_ticket_dia = 'red'
    if ticket_medio >= meta_ticketmedio_ano:
        cor_ticket_ano = 'green'
    else:
        cor_ticket_ano = 'red'

    mail.HTMLBody = f'''
    <p>Bom dia, {nome}</p>

    <p>O resultado de ontem <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da <strong>Loja {loja}</strong> foi:</p>

    <table>
      <tr>
        <th>Indicador</th>
        <th>Valor Dia</th>
        <th>Meta Dia</th>
        <th>Cenário Dia</th>
      </tr>
      <tr>
        <td>Faturamento</td>
        <td style="text-align: center">R${faturamento_dia:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_dia:.2f}</td>
        <td style="text-align: center"><font color="{cor_fat_dia}">◙</font></td>
      </tr>
      <tr>
        <td>Diversidade de Produtos</td>
        <td style="text-align: center">{qtd_produtos_dia}</td>
        <td style="text-align: center">{meta_qtdeprodutos_dia}</td>
        <td style="text-align: center"><font color="{cor_qtd_dia}">◙</font></td>
      </tr>
      <tr>
        <td>Ticket Médio</td>
        <td style="text-align: center">R${ticket_medio_dia:.2f}</td>
        <td style="text-align: center">R${meta_ticketmedio_dia:.2f}</td>
        <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
      </tr>
    </table>
    <br>
    <table>
      <tr>
        <th>Indicador</th>
        <th>Valor Ano</th>
        <th>Meta Ano</th>
        <th>Cenário Ano</th>
      </tr>
      <tr>
        <td>Faturamento</td>
        <td style="text-align: center">R${faturamento_ano:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_ano:.2f}</td>
        <td style="text-align: center"><font color="{cor_fat_ano}">◙</font></td>
      </tr>
      <tr>
        <td>Diversidade de Produtos</td>
        <td style="text-align: center">{qtd_produtos_ano}</td>
        <td style="text-align: center">{meta_qtdeprodutos_ano}</td>
        <td style="text-align: center"><font color="{cor_qtd_ano}">◙</font></td>
      </tr>
      <tr>
        <td>Ticket Médio</td>
        <td style="text-align: center">R${ticket_medio:.2f}</td>
        <td style="text-align: center">R${meta_ticketmedio_ano:.2f}</td>
        <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
      </tr>
    </table>

    <p>Segue em anexo a planilha com todos os dados para mais detalhes.</p>

    <p>Qualquer dúvida estou à disposição.</p>
    <p>Att., </p>
    <p>Antonio Neto </p>
    '''
    
    anexo = pathlib.Path().cwd() / caminho / loja / f'{dia_indicador.day}-{dia_indicador.month}-{loja}.xlsx'
    mail.Attachments.Add(str(anexo))
    mail.Send()
    print('Email enviado para a loja {}'.format(loja))
```

### Passo 7 - Criação dos Rankings das lojas

```python
#criando ranking anual
faturamento_lojas = vendas.groupby('Loja')[['Loja', 'Valor Final']].sum()
faturamento_lojas_ano = faturamento_lojas.sort_values(by='Valor Final', ascending=False)
faturamento_lojas_ano['Valor Final Formatado'] = faturamento_lojas_ano['Valor Final'].apply(lambda x: 'R$ {:,.2f}'.format(x))
ranking_arquivo = r'Ranking anual.xlsx'
faturamento_lojas_ano.to_excel(r'Backup Arquivos Lojas\{}'.format(ranking_arquivo))
display(faturamento_lojas_ano)

#criando ranking diário
vendas_dia = vendas.loc[vendas['Data'] == dia_indicador, :]
faturamento_lojas_dia = vendas_dia.groupby('Loja')[['Loja', 'Valor Final']].sum()
faturamento_lojas_dia = faturamento_lojas_dia.sort_values(by='Valor Final', ascending=False)
faturamento_lojas_dia['Valor Final Formatado'] = faturamento_lojas_dia['Valor Final'].apply(lambda x: 'R$ {:,.2f}'.format(x))
ranking_diario_arquivo = r'{}_{}_Ranking Diário.xlsx'.format(dia_indicador.day, dia_indicador.month)
faturamento_lojas_dia.to_excel(r'Backup Arquivos Lojas\{}'.format(ranking_diario_arquivo))
display(faturamento_lojas_dia)
```

### Passo 8 - Enviar Rankings para diretoria pelo E-mail

```python
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = emails.loc[emails['Loja'] == 'Diretoria', 'E-mail'].values[0]
mail.Subject = 'Ranking Dia {}/{}'.format(dia_indicador.day, dia_indicador.month)
mail.Body = f'''
Prezados, bom dia.

Melhor Loja do Dia em Faturamento: Loja {faturamento_lojas_dia.index[0]} com R$ {faturamento_lojas_dia['Valor Final'].values[0]:,.2f}
Pior Loja do Dia em Faturamento: Loja {faturamento_lojas_dia.index[-1]} com R$ {faturamento_lojas_dia['Valor Final'].values[-1]:,.2f}

Melhor Loja do Ano em Faturamento: Loja {faturamento_lojas_ano.index[0]} com R$ {faturamento_lojas_ano['Valor Final'].values[0]:,.2f}
Pior Loja do Ano em Faturamento: Loja {faturamento_lojas_ano.index[-1]} com R$ {faturamento_lojas_ano['Valor Final'].values[-1]:,.2f}

Segue em anexo o ranking anual e diário de todas as lojas.

Qualquer dúvida estou à disposição.
Att.,
Antonio Neto
'''

anexo = pathlib.Path().cwd() / caminho / f'{dia_indicador.day}_{dia_indicador.month}_Ranking Diário.xlsx'
mail.Attachments.Add(str(anexo))
anexo = pathlib.Path().cwd() / caminho / 'Ranking anual.xlsx'
mail.Attachments.Add(str(anexo))

mail.Send()
print('Email enviado para a Diretoria com os rankings anual e diário')
```
